```python
import time
import numpy as np
import admm
import math
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.datasets import load_boston
```

```python
table = np.loadtxt("results.txt")
table = table.T
p = 5000
lower = -15.0
upper = 1.0
num = 100
lamba_set = []
for t in range(num):
    k = math.exp(upper - (upper - lower)*t/num)
    #print t, k
    lamba_set.append(k)
```

```python
plt.figure()
for i in range(p):
    plt.plot(- np.log(lamba_set), table[i])
plt.savefig('lambda_path.jpg')
plt.show()
```
